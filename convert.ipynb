```python
import yaml
import os

def convert_yaml_to_yolo(yaml_file, output_image_folder, output_label_folder):
    with open(yaml_file, 'r') as file:
        data = yaml.safe_load(file)

    for entry in data:
        img_path = entry['path']
        img_name = os.path.basename(img_path)
        label_file_path = os.path.join(output_label_folder, img_name.replace('.png', '.txt').replace('.jpg', '.txt'))

        with open(label_file_path, 'w') as label_file:
            for box in entry['boxes']:
                label = box['label']
                x_min, y_min, x_max, y_max = box['x_min'], box['y_min'], box['x_max'], box['y_max']
                width = x_max - x_min
                height = y_max - y_min
                x_center = (x_min + x_max) / 2
                y_center = (y_min + y_max) / 2

                # Normalize coordinates (assuming image size is 1280x720)
                x_center /= 1280
                y_center /= 720
                width /= 1280
                height /= 720

                # Convert label to class index (you may need a mapping)
                label_to_class = {'green': 0, 'yellow': 1, 'red': 2, 'off': 3}
                class_id = label_to_class.get(label.lower(), -1)  # Default to -1 if not found

                label_file.write(f"{class_id} {x_center} {y_center} {width} {height}\n")

# Example usage
convert_yaml_to_yolo('dataset_train_rgb/train.yaml', 'dataset/images/train/', 'dataset/labels/train/')
```

```python
import os
import shutil
import yaml

def convert_yaml_to_yolo(yaml_file, image_folder, output_image_folder, output_label_folder, class_mapping):
    # Creazione delle cartelle per immagini e annotazioni
    os.makedirs(output_image_folder, exist_ok=True)
    os.makedirs(output_label_folder, exist_ok=True)

    # Estrai le classi uniche dal mapping per creare 'classes.txt'
    classes = list(class_mapping.values())
    with open(os.path.join(output_image_folder, 'classes.txt'), 'w') as class_file:
        for cls in classes:
            class_file.write(f"{cls}\n")

    with open(yaml_file, 'r') as file:
        data = yaml.safe_load(file)

    for entry in data:
        img_path = entry['path']
        img_name = os.path.basename(img_path)
        
        # Verifica se l'immagine si trova nella cartella delle immagini
        src_image_path = os.path.join(image_folder, img_name)
        
        # Copia l'immagine nella cartella di destinazione
        if os.path.exists(src_image_path):
            shutil.copy(src_image_path, os.path.join(output_image_folder, img_name))

        label_file_path = os.path.join(output_label_folder, img_name.replace('.png', '.txt').replace('.jpg', '.txt'))

        with open(label_file_path, 'w') as label_file:
            for box in entry['boxes']:
                label = box['label']
                x_min, y_min, x_max, y_max = box['x_min'], box['y_min'], box['x_max'], box['y_max']
                width = x_max - x_min
                height = y_max - y_min
                x_center = (x_min + x_max) / 2
                y_center = (y_min + y_max) / 2

                # Normalizza le coordinate
                x_center /= 1280
                y_center /= 720
                width /= 1280
                height /= 720

                # Mappa l'etichetta al suo indice di classe
                class_id = class_mapping.get(label.lower(), -1)  # Se l'etichetta non è trovata, viene dato -1

                label_file.write(f"{class_id} {x_center} {y_center} {width} {height}\n")

# Esempio di utilizzo
class_mapping = {'green': 'green', 'yellow': 'yellow', 'red': 'red', 'off': 'off'}
yaml_file_train = 'c:/Users/Lorenzo Vigorelli/OneDrive - Università degli Studi di Padova/MAGISTRALE UNIPD/Third semester/Computer vision/OBJECT-RECOGNITION/dataset/dataset_train_rgb/train.yaml'
yaml_file_test = 'c:/Users/Lorenzo Vigorelli/OneDrive - Università degli Studi di Padova/MAGISTRALE UNIPD/Third semester/Computer vision/OBJECT-RECOGNITION/dataset/dataset_test_rgb/test.yaml'

image_folder_train = 'c:/Users/Lorenzo Vigorelli/OneDrive - Università degli Studi di Padova/MAGISTRALE UNIPD/Third semester/Computer vision/OBJECT-RECOGNITION/dataset/dataset_train_rgb/rgb/train'
image_folder_test = 'c:/Users/Lorenzo Vigorelli/OneDrive - Università degli Studi di Padova/MAGISTRALE UNIPD/Third semester/Computer vision/OBJECT-RECOGNITION/dataset/dataset_test_rgb/rgb/test'

output_image_folder_train = 'dataset/images/train'
output_label_folder_train = 'dataset/labels/train'

output_image_folder_test = 'dataset/images/test'
output_label_folder_test = 'dataset/labels/test'

convert_yaml_to_yolo(yaml_file_train, image_folder_train, output_image_folder_train, output_label_folder_train, class_mapping)
convert_yaml_to_yolo(yaml_file_test, image_folder_test, output_image_folder_test, output_label_folder_test, class_mapping)
```

```python
import os

# Mappatura delle etichette stringa a numeri
label_map = {
    'green': 0,
    'yellow': 1,
    'red': 2,
    'off': 3,
    '-1': 4
}

# Funzione per sostituire le etichette nei file
def convert_labels_in_file(file_path):
    # Legge il contenuto del file
    with open(file_path, 'r') as file:
        lines = file.readlines()
    
    # Apre il file per la scrittura e sostituisce le etichette
    with open(file_path, 'w') as file:
        for line in lines:
            # Dividi la linea in parti
            parts = line.split()
            
            # Se il primo elemento è una label stringa, lo sostituiamo con il numero
            if parts[0] in label_map:
                parts[0] = str(label_map[parts[0]])  # sostituire la stringa con il numero
            
            # Scrivi la riga modificata nel file
            file.write(" ".join(parts) + "\n")

# Funzione per scansionare una directory e processare tutti i file .txt
def process_labels_in_directory(directory_path):
    # Scansiona la directory per i file .txt
    for filename in os.listdir(directory_path):
        if filename.endswith(".txt"):  # Processa solo i file .txt
            file_path = os.path.join(directory_path, filename)
            convert_labels_in_file(file_path)

# Elenco delle directory contenenti le etichette di allenamento e di test
label_directories = [
    "dataset/labels/train",
    "dataset/labels/test"
]

# Esegui la sostituzione per tutte le directory specificate
for directory in label_directories:
    process_labels_in_directory(directory)

print("Sostituzione delle etichette completata!")
```
